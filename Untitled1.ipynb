```python
import pandas as pd

# Replace 'your_file.xlsx' with your actual Excel file name
file_path = 'questions.xlsx'

# Read the Excel file into a DataFrame
df = pd.read_excel(file_path, header=None)

# Create separate DataFrames for questions and answers
questions = df.iloc[::2].reset_index(drop=True)
answers = df.iloc[1::2].reset_index(drop=True)

# Combine questions and answers into a single DataFrame
result_df = pd.concat([questions, answers], axis=1)
result_df.columns = ['Questions', 'Answers']

# Remove leading numbers from questions
result_df['Questions'] = result_df['Questions'].str.replace(r'^\d+\.\s*', '', regex=True)

# Sort the DataFrame alphabetically based on 'Questions'
result_df_sorted = result_df.sort_values(by='Questions')

display(result_df_sorted)
```
