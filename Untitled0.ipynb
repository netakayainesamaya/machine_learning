```python
input = [1, 2, 3, 6, 9, 10, 12]
results = []

middle = False

for i in range(0, len(input)-1):

  if input[i] + 1 == input[i+1]:

    if not middle:
      start_interval = str(input[i])

    middle = True
  else:
      if middle:
        end_interval = "-" + str(input[i])
        middle = False
        results.append(start_interval+end_interval)
      else:
        results.append(input[i])

  if i==(len(input)-2):
    if middle:
      end_interval = "-" + str(input[i+1])
      middle = False
      results.append(start_interval+end_interval)
    else:
      results.append(input[i+1])

print(results)
```
